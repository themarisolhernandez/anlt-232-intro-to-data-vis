# Simple line plots




Perhaps the simplest of all plots is the visualization of a single function $y = f(x)$. Here
we will take a first look at creating a simple plot of this type. 


## pyplot

Given an array of $x$ values, we can easily plot any numpy function, using the the *pyplot* interface.


```python
# these are done in my startup file
# import matplotlib.pyplot as plt
# import numpy as np

x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
plt.show()
```

If we want to create a single figure with multiple lines, we just call the plot
function multiple times:

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

## The API methods

Alternatively we can explicitly use all the *API* methods. 

We start by creating a `figure` and an `axes`. 

In their simplest form, a figure and axes can be created as follows:

```python
fig = plt.figure()
ax = plt.axes()
plt.show()
```

In matplotlib, the `figure` (an instance of the class `plt.Figure`) can be thought of as a
single container which contains all the *objects* representing `axes, graphics, text, labels`,
etc. The `axes` (an instance of the class `plt.Axes`) is what we see above: a bounding
box with ticks (0 to 1) and labels, which will eventually contain other plot elements. 

It's common to use the variable name `fig` to refer to a `figure` instance, and
`ax` to refer to an `axes` instance or set of `axes` instances.

Once we have created an set of `axes`, we can use the `ax.plot` function to plot some data.

Lets start with a simple sinusoid like above.

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))

plt.show()
```

## Adjusting the Plot: Line *Colors* and *Styles*

### Color

The first adjustment you might wish to make to a plot is to control the line colors and
styles. The `plt.plot()` function takes additional arguments which can be used to
specify these. To adjust the *color*, you can use the `color` keyword, which accepts a
*string* argument representing virtually any imaginable color. The color can be specified
in a variety of ways, which well show below:

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (works for rgb & cmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Greyscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex color code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values between 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all html color names are supported
plt.show()
```

Note that if no color is specified, matplotib will automatically cycle through a set of
default colors for multiple lines.


### Line Styles

Similarly, the *line style* can be adjusted using the `linestyle` keyword:

```python
# Let's make this plot a bit bigger (and stretchable)
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 6, linestyle='-') # solid
plt.plot(x, x + 7, linestyle='--') # dashed
plt.plot(x, x + 8, linestyle='-.') # dashdot
plt.plot(x, x + 9, linestyle=':') # dotted;

plt.show()
```

If you would like to be *extremely* terse, these linestyle codes and color codes can be
combined into a single non-keyword argument to the `plt.plot()` function:

```python
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r') # dotted red

plt.show()
```

These *single-character color codes* reflect the standard abbreviations in the **RGB**
(*Red/Green/Blue*) and **CMYK** (*Cyan/Magenta/Yellow/blacK*) color systems, commonly
used for digital color graphics.

There are many other keyword arguments that can be used to fine-tune the appearance
of the plot: for more details, refer to [matplotibs online documentation](https://matplotlib.org/stable/index.html), or the
docstring of the `plt.plot()` function.

## Adjusting the Plot: Axes limits

Matplotlib does a fairly good job of choosing default *axes limits* for your plot.

But sometimes its nice to have finer control. Here well briefly see how to change the limits
of the $x$ and $y$ axes. The most basic way to do this is to use the `plt.ylim()` methods to set the 
numerical limits of the $x$ and $y$ axes:

```python
plt.plot(x, np.exp(-x/5))
plt.xlim(-1, 11)
plt.ylim(-0.5, 1.5)
plt.show()
```

If for some reason youd like either axis to be displayed in *reverse*, you can simply
reverse the order of the arguments:

```python
plt.plot(x, np.exp(-x/5))
plt.xlim(-1, 11)
plt.ylim(1.5, -0.5)
plt.show()
```

##### plt.axis()

A useful related method is **plt.axis()** (note here the potential confusion between *axes* with an **e**, and *axis* with an **i**). The `plt.axis()` method allows you to set the $x$ and $y$ limits with a *single call*, by passing a *list* which specifies `[xmin, xmax, ymin, ymax]`:

```python
plt.plot(x, np.cos(x)*np.exp(-x/5))
plt.axis([-1, 11, -1.5, 1.5])
plt.show()
```

The `plt.axis()` method goes even beyond this, allowing you to do things like automatically
tighten the bounds around the current plot:

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
plt.show()
```

It allows even higher-level specifications, such as ensuring an equal aspect ratio so
that on your screen, *one unit* in $x$ is equal to *one unit* in $y$:

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
plt.show()
```

Have a look at the `plt.axis` docstring for useful information.

## Labeling things in your plot

**Titles** and **axis labels** are the simplest of these: there are simple methods which can be used to
quickly set these:

```python
plt.plot(x, np.sin(x))
plt.title(r"A Sine Curve = $\sqrt{1 - \cos^2(x)}$")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

The **position, size**, and **style** of these *labels* can be adjusted using optional arguments
to the function. For more information, see docstrings of each of these functions.

----------------------




-----------------------


When *multiple lines* are being shown within a single axes, it can be useful to create a
plot legend that labels each line type. 
Again, matplotlib has a built-in way of quickly creating such a legend. 
It is done via the (you guessed it) `plt.legend()` method.

Though there are several valid ways of using this, I find it easiest to specify the label
of each line using the label keyword of the plot function:

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
plt.show()
```

As you can see, the `plt.legend()` function keeps track of the line style and color, and
*matches* these with the correct label. 

There is a lot you can do with the *legend*. More information on specifying and formatting
plot legends can be found in the....you guessed it: the `plt.legend?` doc string

```python
plt.legend?
```

## Pyplot / API  -  *Gotcha's*

While most **plt** (pyplot) functions translate directly to **ax** (API) methods, such as 

* `plt.plot()`  `ax.plot()`, 
* `plt.legend()`  `ax.legend()`, 
* etc.

this is not the case for all commands.

In particular, functions to **set limits**, **labels**, and **titles** are slightly modified.

For transitioning between **pyplot** (matlab-style) functions and **object-oriented** (API) methods, make the following changes:


| **pyplot**    |   | **OO API**       |
|---------------|---|------------------|
| `plt.xlabel()`|  | `ax.set_xlabel()`|
| `plt.ylabel()`|  | `ax.set_ylabel()`|
| `plt.xlim()`  |  | `ax.set_xlim()`  |
| `plt.ylim()`  |  | `ax.set_ylim()`  |
| `plt.title()` |  | `ax.set_title()` |


Using the **object-oriented API** interface to plotting, rather than calling these functions individually,
it is more convenient to use the `ax.set()` method to set all these properties at once:

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)', title='A Simple Plot')
plt.show()
```
